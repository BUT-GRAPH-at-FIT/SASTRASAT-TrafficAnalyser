```python
import numpy as np
import faiss
import math
from matching.tools import find_matches

d = 128
db_n = 10000
q_n = 100
knn_k = 10
b_size = 10


xb = np.random.random((db_n, d)).astype('float32')
xb = xb / np.linalg.norm(xb, axis=1, keepdims=True)

xq = np.random.random((q_n, d)).astype('float32')
xq = xq / np.linalg.norm(xq, axis=1, keepdims=True)

sims, ids = find_matches(xq, xb, emb_size=d, batch_size=b_size)

print("DS:", sims)
print("Is:", ids)
```

```python
from cuml.neighbors import NearestNeighbors
import cudf
import cupy as cp

import os
# Select a particular GPU to run the notebook
os.environ["CUDA_VISIBLE_DEVICES"]="1" # or replace '1' with which GPU you want to use if you

d = 128
model = NearestNeighbors(n_neighbors=10000, algorithm='ivfflat', metric='cosine')
data = cp.random.random((10000, d), dtype=cp.float32)
query = cp.random.random((100, d), dtype=cp.float32)

batch_size = 2

model.fit(data)

for i in range(0, data.shape[0], batch_size):
    j = min(i + batch_size, data.shape[0])
    distances, indices = model.kneighbors(query[i:j])

print(indices)
print(distances)
```
